```python
import requests 
from bs4 import BeautifulSoup 
 
base_url = "https://scrapeme.live/shop/page" 
pages = range(1, 49) # max page (48) + 1

def extract_details(page): 
	# concatenate page number to base URL 
	response = requests.get(f"{base_url}/{page}/") 
	soup = BeautifulSoup(response.text, "html.parser") 
 
	pokemon_list = [] 
	for pokemon in soup.select(".product"): # loop each product 
		pokemon_list.append({ 
			"id": pokemon.find(class_="add_to_cart_button").get("data-product_id"), 
			"name": pokemon.find("h2").text.strip(), 
			"price": pokemon.find(class_="price").text.strip(), 
			"url": pokemon.find(class_="woocommerce-loop-product__link").get("href"), 
		}) 
	return pokemon_list
```

```python
extract_details()
```

```python
# 서버에서 하나의 IP별로 접속하는 제한이 걸려있을 수도 있고,
# 네트워크 대역폭과 서버 처리량이 한정되어 있기 때문에 
# concurrency에 제한을 둔다.

# native coroutine 이용 asynchronous 

# >= 3.5

import asyncio

max_concurrency = 10

semaphore = asyncio.Semaphore(max_concurrency)

async def demoFunction(i) :
    async with semaphore :
        await asyncio.sleep(1)
        print(f"Hello {i}")

async def main() :
    tasks = [
        demoFunction(i)
        for i in range(100)
    ]
    await asyncio.gather(*tasks)

# ipynb 에서는 사용할 수 없음
# 동기 함수에서 비동기 함수를 실행하기 위해 이벤트 루프 생성 
# loop = anyncio.get_event_loop()
# loop.run_until_complete(main())
# loop.close()

# ipynb는 자체 이벤트루프 위에서 돌기 때문에 await를 쓰면 됨
await main()
```
